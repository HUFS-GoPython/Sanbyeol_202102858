### 3.list comprehension

```python
- 지능형 리스트
- 주어진 리스트를 포함하는 새로운 리스트를 형성하는 것
- list + for_loop 을 한줄로 해결하는 방식
- [A for A' in B] 
   
   - B에 있는 A'에서 A를 가진 리스트를 반환
   - B : 기존의 리스트
   - A' : B의 자료 형태
   - A : 새로운 리스트의 요소
   
- 조건+ :[A for A' in B] if <조건식>
        
        [A for A' in B] if <조건식> and <조건식>
```

```python
[d for d in drinks]
```

```python
[d[0] for d in drinks] #처음만 뽑고싶은것
```
