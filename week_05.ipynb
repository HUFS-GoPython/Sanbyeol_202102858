### while loop

```python
- 어떤 조건이 만족하는 동안 명령 블록을 수행하낟.
- if + for loop
- 해당 조건이 거짓이면 멈춘다 > 언제 멈추는 지 그 지점을 정확히 파악해야함
```

```python
### 둘의 차이
- for : 반복 횟수가 정해짐, 정확함, 변하지 않음
-예 : 성적 산출에서 횟수만 정해져있음

- While : 횟수가 정해지지 않음(중요하지않다) 어떤 조건을 만족시킬때 빠져나오게 하는게 중요할때 사용
예 : 가위바위보
```

```python
for i in range(10):
    print(i)
```

```python
j = 0 # 변수를 설정해야함 #초기화

while j < 3: #0,1,2,False(멈춤)
    print(j)
    j  += 1 #변화
    
    # 위 예시와 똑같은 것.
    # while 변수를 초기화해야하는 이유, 조건을 주기 위해서는 변수가 위에서 선언이 되어야 한다.
```

```python
# 퀴즈
# for loop
abcd = ['a','b','c','d']

for ch in abcd:
    print(ch)
```

```python
# while

j = 0

while j < len(abcd):
    print(abcd[j])
    j += 1
```

```python
### 반복문 제어

- 되도록 사용하지 않는다. 의도치 않은 오류를 위해
- break : 논리적으로 반복을 종료시킬 때
- else : 어떤 조건이 완전히 끝났을때 한번 더 실행함. break시에 실행 안됨
- continue : 특정 조건을 건너뛰고 다음 조건 실행
```

```python
j = 0 
while j < 10:
    print(j)
    if j == 6: break
    j  += 1 #더 이상 밑에 것을 타지않고 멈춤, 6까지는 프린트
```

```python

# 구구단
# 변수는 입력을 받아서 사용함.
- 1. n단 출력
- 2. 1~n 단까지 출력
n > 10보다 클 경우 대비
- 3. n,m단 모두 입력받아서 출력
- 4. 프로그램으로 만듦

 - 메뉴: 1) n단 2) 1-n단, 3) n-m 단
    - 메뉴에 없는 옵션을 선택할때
```

```python
# 1. n단 출력
n = int(input('구하고 싶은 구구단을 입력하세요'))
while True:
    if (n > 9) or (n <1):
        print('올바른 숫자를 입력해주세요.')
        n = int(input('구하고 싶은 구구단을 입력하세요'))
        
    else:
        for i in range(0,9):
            i += 1
            result = n*i
            print(f'{i} X {n} = {result}')
        break
```

```python
# 2. 1~n 단까지 출력

n = int(input('단을 입력하세요:'))
while True:
    if n<=0 or n>9:
        print('올바른 숫자를 입력해주세요.')
        n = int(input('단을 입력하세요:'))
    else:
        for i in range(1,n+1):
            print(f'===={i}단====')
            for j in range(1,10):
                result = j*i
                print(f'{j} X {i} = {result}')
        break
```

```python
#  3. n,m단 모두 입력받아서 출력
```

```python

n = int(input('첫 단을 입력하세요:'))
m = int(input('마지막 단을 입력하세요:'))
while True:
    if n >= m or (n and m >9) or (n and m < 0):
        print('입력 숫자가 잘못 되었습니다. 다시 입력해주세요.')
        n = int(input('첫 단을 입력하세요:'))
        m = int(input('마지막 단을 입력하세요:'))
    
    else:
        for i in range(n,m+1):
            print(f'==={i}단===')
            for j in range(1,10):
                result = j * i
                print(f'{j} X {i} = {result}')
        break
   
```

# Brown
- categories()

```python
brown = nltk.corpus.brown
```

```python
brown.categories()  # brown에 있는 카테고리 리스트
```

```python
brown.words(categories= 'adventure')
```

```python
fdist = nltk.FreqDist([w.lower() for w in brown.words(categories = 'adventure')])
# 카테고리가 adventure인 단어들의 freqdist
```

```python
fdist['men']
# 카테고리가 adventure인 단어들 중 men의 빈도수
```

```python
# modal verbs
modals = ['can', 'could', 'will', 'would', 'may', 'might']

for verb in modals:
    print(verb, fdist[verb])
# 카테고리가 adventure인 단어들 중 조동사들의 빈도수
```

### conditional frequency distribution
- 조건에 따른 fdist


```python
cfd = nltk.ConditionalFreqDist(
    # 각 장르에 따른 frquency distribution을 만든다. 
    (genre, word)
    for genre in brown.categories() # genre는 brown의 카테고리들
    for word in brown.words(categories=genre) # word는 genre에 해당하는 카테고리의 단어들)
```
